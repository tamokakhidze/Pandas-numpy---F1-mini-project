```python
df['dob'] = pd.to_datetime(df['dob']) #dob სვეტი ვაქციოთ დეითთაიმ იბიექტად
```

```python
count = df['dob'].dt.year.value_counts().sort_index() #date of birth სვეტიდან ვაექსტრაქტებთ year-ს, ვითვლით თითოეულის რაოდენობას და ვასორტირებთ
```

```python

top_20_dob_counts = count.nlargest(20) #თავდაპირველად დიაგრამა ყველა მონაცემის მიხედვით ავაგე, მაგრამ იქედან გამომდინარე, რომ მონაცემების რაოდენობა რამ
#დენიმე ასეულია, უკეთესი ხილვადობისთვის 20 ყველაზე დიდ მონაცემზე დაყრდნობით ავაგე საბოლოოდ.
```

```python
plt.figure(figsize=(12, 6)) #დიაგრამას ვანიჭებთ ზომას 
```

```python
top_20_dob_counts.plot(kind='bar') #ვაგებთ ბარს. kind-ში ვუთითებთ თუ რა ტიპის დიაგრამა გვინდა.
```

```python
plt.xlabel('Date of Birth (Year)')
plt.ylabel('Count')
plt.title('Top 20 Date of Birth Bar Chart') 
```
